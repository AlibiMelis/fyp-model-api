```python
df = df.drop(['operation', 'balance', 'k_symbol', 'bank', 'account'], axis=1)
```

```python
def map_function(type):
    if type == 'PRIJEM':
        return 'Income'
    elif type == 'VYDAJ':
        return 'Expense'
    return 'Unknown'
df = df.loc[df['type'] != 'VYBER']
df['type'].value_counts()
```

```python
df['type'] = df['type'].map(map_function)
df['type'].value_counts()
```

```python
df.sample(10)
```

```python
df['account_id'].value_counts()
```

```python
def map_function2(type):
    if type == 'Income':
        return 0
    elif type == 'Expense':
        return 1
    return -1
df['type'] = df['type'].map(map_function2)
df['type'].value_counts()
```

```python
df.sample(10)
```

```python
df['date'] = pd.to_datetime(df['date'], format='%y%m%d')
```

```python
df.sample(10)
```

```python
df.sort_values(by='date', inplace=True)
df.head()
```

```python
def map_function3(amount):
    return amount * 0.36
df['amount'] = df['amount'].map(map_function3)
df.sample(10)
```

```python
df.to_csv('formatted_data.csv', index=False)
```
