```python
dataset = makeDatetimeIndex(single_user_data)

# Step 1
data = dataset.filter(['amount']).values
window = 90 # consider last 90 days

# Step 2
scaler = MinMaxScaler(feature_range=(0, 1))
normalized_data = scaler.fit_transform(data)

# Step 3
training_data_size = math.ceil(data.size * 0.8)
training_data = normalized_data[0:training_data_size]
testing_data = normalized_data[training_data_size - window:]

# Step 4
x_train = [] # features
y_train = [] # labels
x_test = []
y_test = []


for i in range(window, training_data_size):
    x_train.append(training_data[(i - window):i])
    y_train.append(training_data[i])

for i in range(window, len(testing_data)):
    x_test.append(testing_data[(i - window):i])
    y_test.append(testing_data[i])
    
# Step 5
x_train = np.array(x_train)
y_train = np.array(y_train)
x_train = x_train.reshape(x_train.shape[0], x_train.shape[1], 1)

x_test = np.array(x_test)
y_test = np.array(y_test)
x_test = x_test.reshape(x_test.shape[0], x_test.shape[1], 1)
```

```python
plt.figure(figsize=(16, 9))
plt.title('Testing data - Expenses over time')
plt.plot(dataset['amount'][training_data_size:])
plt.xlabel('Date')
plt.ylabel('Stock price (closing) in US$')
plt.show()
```

```python
lstm_model=Sequential()
lstm_model.add(LSTM(units=50,return_sequences=True,input_shape=(x_train.shape[1], 1)))
lstm_model.add(LSTM(units=50))
lstm_model.add(Dense(1))
lstm_model.compile(loss='mean_squared_error',optimizer='adam')
lstm_model.summary()
```

```python
lstm_model.fit(x_train, y_train, batch_size=1, epochs=1, verbose=1)
```

```python
y_predicted = lstm_model.predict(x_test)
y_predicted = scaler.inverse_transform(y_predicted)
```

```python
y_test_transformed = scaler.inverse_transform(y_test)
print('Validation Root Mean Square Error (RMSE):', np.sqrt(mean_squared_error(y_test_transformed, y_predicted)))
```

```python
validation = dataset.filter(['amount'])[training_data_size:]
validation['Prediction'] = y_predicted
validation.head()
```

```python
plt.figure(figsize=(16, 9))
plt.title('Predicted and Actual expenses on testing data')
plt.plot(validation)
plt.legend(['Actual','Prediction'])
plt.savefig('predictions/prediction1.png')
plt.show()
```
